## Sigmoid function

$g(z) = \frac{1}{1+e^{-z}}\tag{1}$
Lojistik regresyon modelinde $z$, genellikle bir doğrusal regresyon modelinin çıktısıdır. Model, girdiyi doğrusal olarak işler ve ardından sigmoid fonksiyonuna verir. Bu sayede çıktı, 0 ile 1 arasında bir değere dönüşür, bu da sınıflandırma problemlerinde olasılık anlamında yorumlanabilir.

```python
import numpy as np
# Matplotlib grafiklerini interaktif hale getirmek için.
%matplotlib widget
import matplotlib.pyplot as plt
```

np.exp(input_array), input_array içindeki her elemanın üstel değerini ( $e^{z}$) hesaplar (`z`).

```python
input_array = np.array([1,2,3])
exp_array = np.exp(input_array)

print("Input to exp:", input_array)
print("Output to exp:", exp_array)
```

```python
def sigmoid(z):
    """
    Compute the sigmoid of z

    Args:
        z (ndarray): A scalar, numpy array of any size.

    Returns:
        g (ndarray): sigmoid(z), with the same shape as z
         
    """
    g=1/(1+ np.exp(-z))

    return g
```

```python
# -10 ile 10 arasında 0.1 adımlarla eşit aralıklı değerler içeren bir dizi oluştur.
z_values = np.arange(-10,10,0.1)

# z_values dizisinin her bir elemanı için sigmoid fonksiyonunu uygula.
sig_z = sigmoid(z_values)

plt.plot(z_values, sig_z)
plt.title("Sigmoid Function")
plt.xlabel("z")
plt.ylabel("sigmoid(z)")
plt.grid()
plt.show()
```

```python
x_array = np.arange(-10,10,2)
sig_array = sigmoid(x_array)

# NumPy dizilerinin ekrana yazdırılma formatını ayarlamak için kullanılır.
np.set_printoptions(precision=3)
print("x   -   sigmoid(y) ")
print(np.c_[x_array,sig_array])
```

```python

fig,ax = plt.subplots(1,1,figsize=(5,3))
ax.plot(x_array, sig_array, c="b")

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
```

```python
# Veri oluşturma
x_array = np.arange(-10, 10, 0.1)
sig_array = sigmoid(x_array)

# fig: Grafiğin genel yapısını tutar.
# ax: Verilerin çizileceği eksenleri tanımlar ve grafikle ilgili detayları yönetir.

fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(x_array, sig_array, linestyle='--', color='purple', marker='o', markersize=3)

# Başlık ve etiketler
ax.set_title("Sigmoid Function", fontsize=14, fontweight='bold')
ax.set_ylabel('sigmoid(z)', fontsize=12)
ax.set_xlabel('z', fontsize=12)

# Arka plan ve grid
ax.set_facecolor('#f0f0f0')
ax.grid(color='gray', linestyle='--', linewidth=0.5)

# Eksen sınırları
ax.set_xlim(-10, 10)
ax.set_ylim(-0.1, 1.1)

ax.legend(['Sigmoid function'])

plt.show()
```
