# Cost Function 

```python
import numpy as np
import matplotlib.pyplot as plt

# Veri oluşturma
np.random.seed(42)
x = np.linspace(0, 10, 100)  # Giriş verileri
y = 2 * x + 1 + np.random.normal(0, 1, size=len(x))  # Çıkış verileri (y = 2x + 1 + gürültü)

# Parametreleri başlatma
w, b = 0.0, 0.0  # Başlangıç değerleri
alpha = 0.01  # Öğrenme oranı
epochs = 1000  # İterasyon sayısı
m = len(x)  # Veri sayısı

# Cost Function'ın takibi için bir liste
cost_history = []

# Gradient Descent döngüsü
for epoch in range(epochs):
   
    y_pred = w * x + b
    # Cost Function hesaplama
    cost = (1 / m) * np.sum((y_pred - y) ** 2)
    cost_history.append(cost)
    # Gradient hesaplama
    dw = (2 / m) * np.sum((y_pred - y) * x)
    db = (2 / m) * np.sum(y_pred - y)
    w -= alpha * dw
    b -= alpha * db

print(f"Optimized Parameters: w = {w:.2f}, b = {b:.2f}")

plt.figure(figsize=(8, 5))
plt.plot(cost_history)
plt.xlabel("Iteration")
plt.ylabel("Cost (MSE)")
plt.title("Cost Function Reduction Over Iterations")
plt.show()

plt.figure(figsize=(8, 5))
plt.scatter(x, y, label="Data")
plt.plot(x, w * x + b, color="red", label=f"Fitted Line: y = {w:.2f}x + {b:.2f}")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression with Gradient Descent")
plt.show()
```
